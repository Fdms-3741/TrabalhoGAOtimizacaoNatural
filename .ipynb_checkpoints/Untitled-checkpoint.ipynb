# Relatórios


```python
import pandas as pd
```

```python
a = pd.read_pickle('2-resultados_geral.pickle')
b = pd.read_pickle('3-resultados_geral.pickle')
c = pd.read_pickle('resultados_geral.pickle')

a = pd.concat([a,b,c])

a
```

```python
out = a[a['Sucesso']==True].groupby(['P','mu','cxpb','mupb','muidxpb','Tipo']).describe().sort_values(by=('N', 'count'),ascending=False)['Geração final','mean'].to_latex()
with open('tabela2.tex','w') as fil:
    fil.write(out)
```

```python
out = a[a['Sucesso']==True].groupby(['P','mu','cxpb','mupb','muidxpb','Tipo']).describe().sort_values(by=('N', 'count'),ascending=False)['MBF','mean'].to_latex()
with open('tabela3.tex','w') as fil:
    fil.write(out)
```
